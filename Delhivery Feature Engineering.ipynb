```python
df = pd.read_csv('delhivery_data.csv', parse_dates=['trip_creation_time', 'od_start_time', 'od_end_time'])
df.head()
```

- unknown fields = ['is_cutoff', 'cutoff_factor', 'cutoff_timestamp', 'factor', 'segment_factor']

```python
df.drop(['is_cutoff', 'cutoff_factor', 'cutoff_timestamp', 'factor', 'segment_factor'], axis = 1, inplace=True)
df.info()
```

- Source and destination names have some null values

```python
df[pd.isnull(df.source_name) | pd.isnull(df.destination_name)]
```

- We will drop all rows which have either source name null or destination name null

```python
len(df[pd.notnull(df.source_name) & pd.notnull(df.destination_name)])
```

```python
df = df[pd.notnull(df.source_name) & pd.notnull(df.destination_name)]
```

```python
df.head(60)
```

- Grouping by ['trip_uuid', 'source_center', 'destination_center']

```python
grouped_1 = df.groupby(['trip_uuid', 'source_center', 'destination_center']).agg(
    trip_creation_time=pd.NamedAgg(column="trip_creation_time", aggfunc="first"),
    route_type=pd.NamedAgg(column="route_type", aggfunc="first"),
    source_name=pd.NamedAgg(column="source_name", aggfunc="first"),
    destination_name=pd.NamedAgg(column="destination_name", aggfunc="first"),
    od_start_time=pd.NamedAgg(column="od_start_time", aggfunc="first"),
    od_end_time=pd.NamedAgg(column="od_end_time", aggfunc="first"),
    start_scan_to_end_scan=pd.NamedAgg(column="start_scan_to_end_scan", aggfunc="first"),
    actual_distance_to_destination=pd.NamedAgg(column="actual_distance_to_destination", aggfunc="last"),
    actual_time=pd.NamedAgg(column="actual_time", aggfunc="last"),
    osrm_time=pd.NamedAgg(column="osrm_time", aggfunc="last"),
    osrm_distance=pd.NamedAgg(column="osrm_distance", aggfunc="last"),
    actual_time_segmented_sum=pd.NamedAgg(column="segment_actual_time", aggfunc="sum"),
    osrm_time_segmented_sum=pd.NamedAgg(column="segment_osrm_time", aggfunc="sum"),
    osrm_distance_segmented_sum=pd.NamedAgg(column="segment_osrm_distance", aggfunc="sum")
    ).sort_values(by=['trip_uuid', 'od_start_time'])

# grouped_1.loc[grouped_1.source_name == 'Anand_VUNagar_DC (Gujarat)']
grouped_1.loc[['trip-153741093647649320']]
```

```python
grouped_2 = grouped_1.groupby(['trip_uuid']).agg(
    trip_creation_time=pd.NamedAgg(column="trip_creation_time", aggfunc="first"),
    route_type=pd.NamedAgg(column="route_type", aggfunc="first"),
    source_name=pd.NamedAgg(column="source_name", aggfunc="first"),
    destination_name=pd.NamedAgg(column="destination_name", aggfunc="last"),
    od_start_time=pd.NamedAgg(column="od_start_time", aggfunc="first"),
    od_end_time=pd.NamedAgg(column="od_end_time", aggfunc="last"),
    start_scan_to_end_scan=pd.NamedAgg(column="start_scan_to_end_scan", aggfunc="sum"),
    actual_distance_to_destination=pd.NamedAgg(column="actual_distance_to_destination", aggfunc="sum"),
    actual_time=pd.NamedAgg(column="actual_time", aggfunc="sum"),
    osrm_time=pd.NamedAgg(column="osrm_time", aggfunc="sum"),
    osrm_distance=pd.NamedAgg(column="osrm_distance", aggfunc="sum"),
    actual_time_segmented_sum=pd.NamedAgg(column="actual_time_segmented_sum", aggfunc="sum"),
    osrm_time_segmented_sum=pd.NamedAgg(column="osrm_time_segmented_sum", aggfunc="sum"),
    osrm_distance_segmented_sum=pd.NamedAgg(column="osrm_distance_segmented_sum", aggfunc="sum"))
grouped_2.loc[['trip-153741093647649320']]
```

```python
grouped_2['source_state'] = grouped_2['source_name'].str.split('(').str[1].str.slice(0,-1)
grouped_2[['source_city', 'source_place', 'source_code']] = grouped_2['source_name'].str.split('(').str[0].str.strip().str.split('_', 2, expand=True)
grouped_2['destination_state'] = grouped_2['destination_name'].str.split('(').str[1].str.slice(0,-1)
grouped_2[['destination_city', 'destination_place', 'destination_code']] = grouped_2['destination_name'].str.split('(').str[0].str.strip().str.split('_', 2, expand=True)
```

```python
grouped_2.head(20)
```

```python
grouped_2['trip_creation_day'] = grouped_2.trip_creation_time.dt.day
grouped_2['trip_creation_month'] = grouped_2.trip_creation_time.dt.month
grouped_2['trip_creation_year'] = grouped_2.trip_creation_time.dt.year

grouped_2.head(10)
```

```python
grouped_2['od_time'] = (grouped_2.od_end_time - grouped_2.od_start_time).dt.seconds/60
grouped_2.drop(['od_start_time', 'od_end_time'], axis = 1, inplace=True)
grouped_2.head(10)
```
